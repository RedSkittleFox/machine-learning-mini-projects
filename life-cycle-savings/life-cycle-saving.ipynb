```r
library(purrr)

hist_list <- df %>%
    select_if(is.numeric) %>%
    colnames() %>%
    lapply(function(k) 
            ggplot(df, aes_string(x = k)) + 
            xlab( k ) +
            geom_histogram( bins = 8) 
    ) 

grid.arrange(grobs = hist_list, ncol=3)
```
